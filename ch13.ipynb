## ch13_슬라이싱보다는 나머지를 모두 잡아내는 언패킹 사용하라

```python
# 에러 발생
car_ages = [0, 9, 4, 8, 7, 20, 19, 1, 5, 15]
car_ages_descending = sorted(car_ages, reverse=True)
oldest, second_oldest = car_ages_descending
```

```python
# 인덱스와 슬라이싱 사용
oldest = car_ages_descending[0]
second_oldest = car_ages_descending[1]
others = car_ages_descending[2:]
print(oldest, second_oldest, others)
```

```python
# 별표 식 사용
oldest, second_oldest, *others = car_ages_descending
print(oldest, second_oldest, others)
```

```python
# 별표 식 사용 2
oldest, *others, youngest = car_ages_descending
print(oldest, others, youngest)

*others, second_youngest, youngest = car_ages_descending
print(others, second_youngest, youngest)
```

```python
# 별표 식 에러 - 필수가 하나 이상 필요
*others = car_ages_descending
```

```python
# 별표 식 에러 - 하나의 언패킹 패턴에 별표 식 두 개 사용 불가
first, *middle, *second_middle, last = [1, 2, 3, 4]
```

```python
# 별표 식은 항상 list 인스턴스가 됨
short_list = [1, 2]
first, second, *rest = short_list
print(first, second, rest)
```
