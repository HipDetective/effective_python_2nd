```python
# 가능예시지만 실사용은 비추천
favorite_snacks = {
    '짭조름한 과자': ('프레즐', 100),
    '달콤한 과자': ('쿠키', 100),
    '채소': ('당근', 20),
}

(
    (type1, (name1, cals1)),
    (type2, (name2, cals2)),
    (type3, (name3, cals3))
) = favorite_snacks.items()

print(f'제일 좋아하는 {type1}는 {name1}, {cals1} 칼로리입니다.')
print(f'제일 좋아하는 {type2}는 {name2}, {cals3} 칼로리입니다.')
print(f'제일 좋아하는 {type3}는 {name3}, {cals3} 칼로리입니다.')
```

```python
# 언패킹을 이용하여 값을 맞바꾸기 예시 - 언패킹 미사용 케이스
def bubble_sort(a):
    for _ in range(len(a)):
        for i in range(1, len(a)):
            if a[i] < a[i-1]:
                temp = a[i]
                a[i] = a[i-1]
                a[i-1] = temp

names = ['프레즐', '당근', '쑥갓', '베이컨']
bubble_sort(names)
print(names)
```

```python
# 언패킹을 이용하여 값을 맞바꾸기 예시 - 언패킹 사용 케이스
def bubble_sort(a):
    for _ in range(len(a)):
        for i in range(1, len(a)):
            if a[i] < a[i-1]:
                a[i-1], a[i] = a[i], a[i-1]  # 맞바꾸기

names = ['프레즐', '당근', '쑥갓', '베이컨']
bubble_sort(names)
print(names)
```

```python
# 언패킹 미사용 케이스 - 리스트 이터레이션
snacks = [('베이컨', 350), ('도넛', 240), ('머핀', 190)]

for i in range(len(snacks)):
    item = snacks[i]
    name = item[0]
    calories = item[1]
    print(f'#{i+1}: {name}은 {calories} 칼로리입니다.')
```

```python
# 언패킹 사용 케이스 - 더 파이써닉한 코드
snacks = [('베이컨', 350), ('도넛', 240), ('머핀', 190)]

for rank, (name, calories) in enumerate(snacks, 1):
    print(f'#{rank}: {name}은 {calories} 칼로리입니다.')
```
