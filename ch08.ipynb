## ch-08 여러 리터레이터에 대해 나란히 루프를 수행하려면 zip을 사용하라.

```python
names = ['Cecilia', '남궁민수', '毛泽东']
counts = [len(n) for n in names]
print(counts)
```

```python
# 두 리스트 이터레이션 - 1. 리스트의 길이를 사용해 이터레이션
names = ['Cecilia', '남궁민수', '毛泽东']
counts = [len(n) for n in names]

longest_name = None
max_count = 0

for i in range(len(names)):
    count = counts[i]
    if count > max_count:
        longest_name = names[i]
        max_count = count

print(longest_name)
```

```python
# 두 리스트 이터레이션 - 2. enumerate 사용
names = ['Cecilia', '남궁민수', '毛泽东']
counts = [len(n) for n in names]

longest_name = None
max_count = 0

for i, name in enumerate(names):
    count = counts[i]
    if count > max_count:
        longest_name = names[i]
        max_count = count

print(longest_name)
```

```python
# 두 리스트 이터레이션 - 3. zip 내장 함수 사용
names = ['Cecilia', '남궁민수', '毛泽东']
counts = [len(n) for n in names]

longest_name = None
max_count = 0

for name, count in zip(names, counts):
    if count > max_count:
        longest_name = name
        max_count = count

print(longest_name)
```

```python
# zip 사용 - 입력 이터레이터의 길이가 다를 때
names = ['Cecilia', '남궁민수', '毛泽东']
counts = [len(n) for n in names]

names.append('Rosalind')

for name, count in zip(names, counts):
    print(name)
```

```python
import itertools

names = ['Cecilia', '남궁민수', '毛泽东']
counts = [len(n) for n in names]

names.append('Rosalind')

for name, count in itertools.zip_longest(names, counts):
    print(f'{name}: {count}')
```
