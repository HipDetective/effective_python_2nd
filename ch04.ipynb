### C 스타일 형식 문자열의 세 번째 문제점: 값 여러번 사용 시 튜플 내 반복 필요

```python
template = '%s는 음식을 좋아해. %s가 요리하는 모습을 봐요.'
name = '철수'
formatted = template % (name, name)
print(formatted)
```

```python
name = '영희'
formatted = template % (name.title(), name.title())
print(formatted)
```

```python
key = 'my_var'
value = 1.234

old_way = '%-10s = %.2f' % (key, value)
print(old_way)

new_way = '%(key)-10s = %(value).2f' % {'key': key, 'value': value} # 원래 방식
print(new_way)
reordered = '%(key)-10s = %(value).2f' % {'value': value, 'key': key} # 바꾼 방식
print(reordered)

assert old_way == new_way == reordered
```

```python
name = '철수'
template = '%s는 음식을 좋아해. %s가 요리하는 모습을 봐요.'

before = template % (name, name)  # 튜플
print(before)

template = '%(name)s는 음식을 좋아해. %(name)s가 요리하는 모습을 봐요.'
after = template % {'name': name}  # 딕셔너리
print(after)

assert before == after
```

### C 스타일 형식 문자열의 네 번째 문제점: 형식화 식에 딕셔너리을 사용하면 문장이 번잡스러워짐

```python
for i, (item, count) in enumerate(pantry):
    before = '#%d: %-10s = %d' % (
        i + 1,
        item.title(),
        round(count))
    print(before)
    
    after = '#%(loop)d: %(item)-10s = %(count)d' % {
        'loop': i + 1,
        'item': item.title(),
        'count': round(count)
    }
    print(after)

assert before == after
```

```python
pantry = [
    ('아보카도', 1.25),
    ('바나나', 2.5),
    ('체리', 15),
]

for i, (item, count) in enumerate(pantry):
    old_style = '#%d: %-10s = %d' % (
        i + 1,
        item.title(),
        round(count)
    )
    print(old_style)

    new_style = '#{}: {:<10s} = {}'.format(
        i + 1,
        item.title(),
        round(count)
    )
    print(new_style)

    assert old_style == new_style
```

```python
old_template = (
    'Today\'s soup is %(soup)s, '
    'buy one get two %(oyster)s oysters, '
    'and our special entrée is %(special)s.'
)

old_formatted = old_template % {
    'soup': 'lentil',
    'oyster': 'tongyoung',
    'special': 'schnitzel',
}

print(old_formatted)

new_template = (
    'Today\'s soup is {soup}, '
    'buy one get two {oyster} oysters, '
    'and our special entrée is {special}.'
)

new_formatted = new_template.format(
    soup = 'lentil',
    oyster = 'tongyoung',
    special = 'schnitzel',
)

print(new_formatted)

assert old_formatted == new_formatted
```

### 인터폴레이션을 통한 형식 문자열

```python
key = 'my_var'
value = 1.234

formatted = f'{key} = {value}'
print(formatted)
```

```python
key = 'my_var'
value = 1.234

formatted = f'{key!r:<10} = {value:.2f}'
print(formatted)
```

```python
key = 'my_var'
value = 1.234

f_string = f'{key:<10} = {value:.2f}'
c_tuple  = '%-10s = %.2f' % (key, value)
str_arg  = '{:<10} = {:.2f}'.format(key, value)
str_kw   = '{key:<10} = {value:.2f}'.format(key=key, value=value)
c_dict   = '%(key)-10s = %(value).2f' % {'key': key, 'value': value}

assert c_tuple == c_dict == f_string
assert str_arg == str_kw == f_string
```

```python
pantry = [
    ('아보카도', 1.25),
    ('바나나', 2.5),
    ('체리', 15),
]

for i, (item, count) in enumerate(pantry):
    old_style = '#%d: %-10s = %d' % (
        i + 1,
        item.title(),
        round(count)
    )
    print(old_style)

    new_style = '#{}: {:<10s} = {}'.format(
        i + 1,
        item.title(),
        round(count)
    )
    print(new_style)

    f_string = f'#{i+1}: {item.title():<10s} = {round(count)}'
    print(f_string)

    assert old_style == new_style == f_string
```

```python
pantry = [
    ('아보카도', 1.25),
    ('바나나', 2.5),
    ('체리', 15),
]

for i, (item, count) in enumerate(pantry):
    print(f'#{i+1}: '
          f'{item.title():<10s} = '
          f'{round(count)}'
          )
```
