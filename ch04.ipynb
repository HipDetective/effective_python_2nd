### C 스타일 형식 문자열의 첫 번째 문제점: 순서에 따른 오류 발생 가능성

```python
key = 'my_var'
value = 1.234
formatted = '%-10s = %.2f' % (key, value)
print(formatted)
```

```python
key = 'my_var'
value = 1.234
reordered_tuple = '%-10s = %.2f' % (value, key)
print(reordered_tuple)
```

```python
key = 'my_var'
value = 1.234
reordered_string = '%.2f = %-10s' % (key, value)
print(reordered_string)
```

### C 스타일 형식 문자열의 두 번째 문제점: 값을 살짝 변경한다면 가독성이 떨어짐

```python
pantry = [
    ('아보카도', 1.25),
    ('바나나', 2.5),
    ('체리', 15),
]

for i, (item, count) in enumerate(pantry):
    print('#%d: %-10s = %.2f' % (i, item, count))
```

```python
pantry = [
    ('아보카도', 1.25),
    ('바나나', 2.5),
    ('체리', 15),
]

for i, (item, count) in enumerate(pantry):
    print('#%d: %-10s = %.2f' % (
        i + 1,
        item.title(),
        round(count))
    )
```

### C 스타일 형식 문자열의 세 번째 문제점: 값 여러번 사용 시 튜플 내 반복 필요

```python
template = '%s는 음식을 좋아해. %s가 요리하는 모습을 봐요.'
name = '철수'
formatted = template % (name, name)
print(formatted)
```

```python
name = '영희'
formatted = template % (name.title(), name.title())
print(formatted)
```
