## ch07_range보다는 enumerate를 사용하라

```python
from random import randint

random_bits = 0
for i in range(32):
    if randint(0, 1):
        random_bits |= (1 << i)

print(random_bits)
print(bin(random_bits))
```

```python
flavor_list = ['바닐라', '초콜릿', '피칸', '딸기']

for flavor in flavor_list:
    print(f'{flavor} 맛있어요.')
```

```python
# 리스트를 이터레이션하면서 리스트의 몇 번째 원소를 처리 중인지 알아야 할 때 - range(len()) 사용
flavor_list = ['바닐라', '초콜릿', '피칸', '딸기']

for i in range(len(flavor_list)):
    flavor = flavor_list[i]
    print(f'{i + 1}: {flavor}')
```

```python
flavor_list = ['바닐라', '초콜릿', '피칸', '딸기']

it = enumerate(flavor_list)
print(next(it))
print(next(it))
```

```python
# 리스트를 이터레이션하면서 리스트의 몇 번째 원소를 처리 중인지 알아야 할 때 - enumerate() 사용 1
flavor_list = ['바닐라', '초콜릿', '피칸', '딸기']

for i, flavor in enumerate(flavor_list):
    print(f'{i + 1}: {flavor}')
```

```python
# 리스트를 이터레이션하면서 리스트의 몇 번째 원소를 처리 중인지 알아야 할 때 - enumerate() 사용 2
flavor_list = ['바닐라', '초콜릿', '피칸', '딸기']

for i, flavor in enumerate(flavor_list, 1):
    print(f'{i}: {flavor}')
```
