```python
## ch12_스트라이드와 슬라이스를 한 식에 함께 사용하지 말라.
```

```python
x =['빨강', '주황', '노랑', '초록', '파랑', '자주']
odds = x[::2]
evens = x[1::2]
print(odds)
print(evens)
```

```python
# 버그
x = b'mongoose'
y = x[::-1]
print(y)
```

```python
# 유니코드 문자열 스트라이딩
x = '寿司'
y = x[::-1]
print(y)
```

```python
# 에러
w = '寿司'
x = w.encode('utf-8')
y = x[::-1]
z = y.decode('utf-8')
```

```python
x = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
print(x[::2])
print(x[::-2])
```

```python
x = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
print(x[2::2])
print(x[-2::-2])
print(x[-2:2:-2])
print(x[2:2:-2])
```

```python
x = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
y = x[::2]
z = y[1:-1]
print(y)
print(z)
```
