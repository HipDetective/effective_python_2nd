## ch14_복잡한 기준을 사용해 정렬할 때는 key 파라미터를 사용하라

```python
numbers = [93, 86, 11, 68, 70]
numbers.sort()
print(numbers)
```

```python
class Tool:
    def __init__(self, name, weight):
        self.name = name
        self.weight = weight
    
    def __repr__(self):
        return f'Tool({self.name!r}, {self.weight})'

tools = [
    Tool('수준계', 3.5),
    Tool('해머', 1.25),
    Tool('스크류드라이버', 0.5),
    Tool('끌', 0.25),
]

print(tools)
```

```python
# 에러 발생
tools.sort()
```

```python
# key를 사용하여 정렬 - name 기준
print('미정렬:', repr(tools))
tools.sort(key=lambda x: x.name)
print('정렬:', repr(tools))
```

```python
# key를 사용하여 정렬 - weight 기준
print('미정렬:', repr(tools))
tools.sort(key=lambda x: x.weight)
print('정렬:', repr(tools))
```

```python
# key를 사용하여 정렬 - 문자열을 소문자로 변환
places = ['home', 'work', 'New York', 'Paris']
places.sort()
print(f'대소문자 구분: {places}')
places.sort(key=lambda x: x.lower())
print(f'대소문자 무시: {places}')
```

```python
power_tools = [
    Tool('드릴', 4),
    Tool('원형 톱', 5),
    Tool('착암기', 40),
    Tool('연마기', 4),
]
```

```python
saw = (5, '원형 톱')
jackhammer = (40, '착암기')
assert not (jackhammer < saw)
```

```python
drill = (4, '드릴')
sander = (4, '연마기')
assert drill[0] == sander[0]  # 무게 일치
assert drill[1] < sander[1]  # 알파벳 순으로 볼 때 더 작음
assert drill < sander  # 그러므로 드릴이 더 먼저임
```

```python
power_tools.sort(key=lambda x: (x.weight, x.name))
print(power_tools)
```

```python
power_tools.sort(key=lambda x: (x.weight, x.name), reverse=True)
print(power_tools)
```

```python
power_tools.sort(key=lambda x: (-x.weight, x.name), reverse=True)
print(power_tools)
```

```python
# 에러 발생
power_tools.sort(key=lambda x: (x.weight, -x.name), reverse=True)
print(power_tools)
```

```python
power_tools.sort(key=lambda x: x.name)  # name 기준 오름차순
power_tools.sort(key=lambda x: x.weight, reverse=True)  # weight 기준 내림차순
print(power_tools)
```

```python
power_tools.sort(key=lambda x: x.name)  # name 기준 오름차순
print(power_tools)
```

```python
power_tools.sort(key=lambda x: x.weight, reverse=True)  # weight 기준 내림차순
print(power_tools)
```
