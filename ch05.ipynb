## ch05_복잡한 식을 쓰는 대신 도우미 함수를 작성해라.

```python
from urllib.parse import parse_qs

my_value = parse_qs('빨강=5&파랑=0&초록=', keep_blank_values=True)
print(repr(my_value))
```

```python
print('빨강:', my_value.get('빨강'))
print('빨강:', my_value.get('파랑'))
print('빨강:', my_value.get('투명도'))
```

```python
# 질의 문자열이 '빨강=5&파랑=0&초록='인 경우
red = my_value.get('빨강', [''])[0] or 0
green = my_value.get('초록', [''])[0] or 0
opacity = my_value.get('투명도', [''])[0] or 0

print(f'빨강: {red!r}')
print(f'초록: {green!r}')
print(f'투명도: {opacity!r}')
```

```python
# 삼항 조건식 적용
red_str = my_value.get('빨강', [''])
red = int(red_str[0]) if red_str[0] else 0
```

```python
green_str = my_value.get('초록', [''])
if green_str[0]:
    green = int(green_str[0])
else:
    green = 0
```

```python
# 반복 적용 시 도우미 함수를 작성
def get_first_int(values, key, default=0):
    found = values.get(key, [''])
    if found[0]:
        return int(found[0])
    else:
        default
```

```python
# 도우미 함수를 사용한 코드
green = get_first_int(my_value, '초록')
```
