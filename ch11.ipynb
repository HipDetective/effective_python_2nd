## ch11_시퀀스를 슬라이싱 하는 방법을 익혀라.

```python
a = [i for i in 'abcdefgh']
print(a)
print('가운데 2개:', a[3:5])
print('마지막을 제외한 나머지:', a[0:7])
```

```python
assert a[:5] == a[0:5]
```

```python
assert a[5:] == a[5:len(a)]
```

```python
print(a[:])
print(a[:5])
print(a[:-1])
print(a[4:])
print(a[-3:])
print(a[2:5])
print(a[2:-1])
print(a[-3:-1])
```

```python
first_twenty_items = a[:20]
last_twenty_items = a[-20:]

print(first_twenty_items)
print(last_twenty_items)
```

```python
# 에러 발생
print(a[20])
```

```python
b = a[3:]
print('이전:', b)
b[1] = 99
print('이후:', b)
print('변화 없음:', a)
```

```python
print('이전:', a)
a[2:7] = [99, 22, 14]
print('이후:', a)
```

```python
print('이전:', a)
a[2:3] = [47, 11]
print('이후:', a)
```

```python
b = a[:]
assert b == a and b is not a
```

```python
b = a
print('이전 a:', a)
print('이전 b:', b)
a[:] = [101, 102, 103]
assert a is b
print('이후 a:', a)
print('이후 b:', b)
```
